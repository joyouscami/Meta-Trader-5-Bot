```python
import pandas as pd
# Define the symbol and timeframe
symbol = "EURUSD.m"         # Change to any symbol available in your MT5
timeframe = mt5.TIMEFRAME_H1  # Options: M1, M5, M15, M30, H1, H4, D1, etc.

# Define date range: last 30 days
end_time = datetime.now()
start_time = end_time - timedelta(days=30)
```

```python
symbols = mt5.symbols_get()
for s in symbols:
    print(s.name)
```

```python
# Get historical OHLC data
rates = mt5.copy_rates_range(symbol, timeframe, start_time, end_time)

if rates is None or len(rates) == 0:
    print("No market data found.")
    
# Convert to DataFrame
market_df = pd.DataFrame(rates)

# Convert time from timestamp to datetime
market_df['time'] = pd.to_datetime(market_df['time'], unit='s')

# Preview the data
print(market_df.head())
```
